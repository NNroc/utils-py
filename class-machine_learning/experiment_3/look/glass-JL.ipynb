```python
# 导入相关数据库

import pandas as pd
from collections import Counter
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import RepeatedStratifiedKFold, cross_val_score
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, classification_report, accuracy_score,adjusted_rand_score
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering, AffinityPropagation
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# 导入数据集

glass_data= pd.read_csv('glass.data',header=None,delimiter=",",skiprows=1)
glass_test=pd.read_csv('glass.test',header=None,delimiter=",",skiprows=1)

# 合并数据集

glass_muti=pd.concat([glass_data,glass_test])


#插入第一列为缺失的Id，从1开始，到最后一行结束

glass_data.insert(0, 'index', range(1, len(glass_data) + 1))
glass_test.insert(0, 'index', range(1, len(glass_test) + 1))
glass_muti.insert(0, 'index', range(1, len(glass_muti) + 1))
glass_data.columns = ['Id', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type']
glass_test.columns = ['Id', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type']
glass_muti.columns = ['Id', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type']

# 去除 Type 列中的点

glass_muti['Type'] = glass_muti['Type'].str.rstrip('.')
glass_data['Type'] = glass_data['Type'].str.rstrip('.')
glass_test['Type'] = glass_test['Type'].str.rstrip('.')


#将窗玻璃类和非窗玻璃类分别用1和0表示,二分类
# 多分类

type_mapping = {
    ' building_windows_float_processed': 1,
    ' building_windows_non_float_processed': 2,
    ' vehicle_windows_float_processed': 3,
    ' vehicle_windows_non_float_processed': 4,
    ' containers': 5,
    ' tableware': 6,
    ' headlamps': 7
}

# 转换 Type 列的值

glass_muti['Type'] = glass_muti['Type'].map(type_mapping)
# glass_data['Type'] = glass_data['Type'].apply(lambda x: 1 if x in window_glass_types else 0)
glass_data['Type'] = glass_data['Type'].map(type_mapping)
glass_test['Type'] = glass_test['Type'].map(type_mapping)

print(glass_muti.head())
summary = glass_muti.describe()

# 去掉Id列

ncol1 = len(glass_muti.columns)
ncol1_data = len(glass_data.columns)
ncol1_test = len(glass_test.columns)
glass_muti= glass_muti.iloc[:, 1:ncol1]
glass_test = glass_test.iloc[:, 1:ncol1_test]
glass_data = glass_data.iloc[:, 1:ncol1_data]

glassNormalized = glass_muti.iloc[:, 1:ncol1]
ncol2 = len(glassNormalized.columns)
summary = glassNormalized.describe()
print(summary)

# 归一化

for i in range(ncol2):
    mean = summary.iloc[1, i]
    sd = summary.iloc[2, i]
    glassNormalized.iloc[:,i:(i + 1)] = (glassNormalized.iloc[:,i:(i + 1)] - mean) / sd

# 绘制直方图
# 调整子图间距

glassNormalized.hist()
plt.subplots_adjust(hspace=1, wspace=1)
plt.show()

# 对数据进行K倍交叉验证
# 加载数据集

def load_dataset(data):

    X=data.drop(columns=['Type'])
    y=data['Type']
    scaler = StandardScaler()
    X_scaled =scaler.fit_transform(X)
    return X_scaled, y

# 评估模型

def evaluate_model(X, y, model):
    cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=1)
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    return scores

# 测试模型

def get_models():
    models, names = list(),list()
    
    # KMeans聚类
    
    models.append(KMeans(n_clusters=10, random_state=42))
    names.append('KMeans(cluster=10)')
    models.append(KMeans(n_clusters=6, random_state=42))
    names.append('KMeans(cluster=6)')
    models.append(KMeans(n_clusters=4, random_state=42))
    names.append('KMeans(cluster=4)')
    models.append(KMeans(n_clusters=2, random_state=42))
    names.append('KMeans(cluster=2)')

    # DBSCAN

    models.append(DBSCAN(eps=0.5, min_samples=5))
    names.append('DBSCAN(min_samples=5)')
    models.append(DBSCAN(eps=0.5, min_samples=10))
    names.append('DBSCAN(min_samples=10)')
    models.append(DBSCAN(eps=0.5, min_samples=15))
    names.append('DBSCAN(min_samples=15)')

    # 层次聚类

    models.append(AgglomerativeClustering())
    names.append('层次聚类(default)')
    
    # AP聚类

    models.append(AffinityPropagation())
    names.append('AP(default)')

    # 二分类模型
    
    models.append(DecisionTreeClassifier(random_state=42))
    names.append('Tree (Binary)')
    models.append(GaussianNB())
    names.append('Bayes (Binary)')
    models.append(KNeighborsClassifier(n_neighbors=5))
    names.append('KNN (Binary)')

    # 多分类模型
    
    models.append(LogisticRegression(multi_class='multinomial', solver='lbfgs'))
    names.append('Logistic Regression (Multi-class)')
    models.append(RandomForestClassifier(n_estimators=100, random_state=42))
    names.append('Random Forest (Multi-class)')

    return models  , names
    
# 二分类任务
# 划分训练集和测试集

X_train, y_train = load_dataset(glass_data)
X_test,y_test=load_dataset(glass_test)
y_binary_train = ( y_train> 4).astype(int)
y_binary_test = (y_test > 4).astype(int)
models, names = get_models()
for model, name in zip(models, names):
    if 'Binary' in name:
        model.fit(X_train, y_binary_train)
        y_pred = model.predict(X_test)
        print(f"{name} 二分类结果:")
        print(classification_report(y_binary_test, y_pred))
        print("准确率:", accuracy_score(y_binary_test, y_pred))
        print("-" * 50)
# 多分类

for model, name in zip(models, names):
    if 'Multi-class' in name:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        print(f"{name} 多分类结果:")
        print(classification_report(y_test, y_pred))
        print("准确率:", accuracy_score(y_test, y_pred))
        print("-" * 50)

#聚类

# print(glass_muti.head())

X2, y2 = load_dataset(glass_muti)

#调参后聚类

for model, name in zip(models, names):
    if  'cluster'  in name:
        labels = model.fit_predict(X2)
        score = silhouette_score(X2, labels)
        score2=adjusted_rand_score(y2, labels)
        print(f"{name} Silhouette Score: {score:.3f}")
        print(f"{name} Adjusted Rand Score: {score2:.3f}")

for model, name in zip(models, names):
    if  'min_samples' in name:
        labels = model.fit_predict(X2)
        score = silhouette_score(X2, labels)
        score2=adjusted_rand_score(y2, labels)
        print(f"{name} Silhouette Score: {score:.3f}")
        print(f"{name} Adjusted Rand Score: {score2:.3f}")

for model, name in zip(models, names):
    if  'default' in name:
        labels = model.fit_predict(X2)
        score = silhouette_score(X2, labels)
        score2=adjusted_rand_score(y2, labels)
        print(f"{name} Silhouette Score: {score:.3f}")
        print(f"{name} Adjusted Rand Score: {score2:.3f}")
```
